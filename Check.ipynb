Color Space in Opencv 

```python
img = cv2.imread('data/input/img/baboon.png')
#format RGB
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
cv2.imshow('RGB', img)
cv2.waitKey(0)

#format HSV
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow('RGB', img)
cv2.waitKey(0)
2
#format HLS
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
cv2.imshow('RGB', img)
cv2.waitKey(0)

# image into gray
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('gray', img)
cv2.waitKey(0)
```

Menyatukan 2 gambar

```python
img = cv2.imread('data/input/img/baboon.png')
print('Image 1')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

img2 = cv2.imread('data/input/img/lenna.png')
print('Image 2')
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()

#buat menjadi ukurannya sama
img = cv2.resize(img,(400,400))
img2 = cv2.resize(img2,(400,400))

#Blending
# Alpha = 0.7
# Beta = 0.1

blended = cv2.addWeighted(src1=img, alpha=0.7, src2=img2, beta=0.1, gamma=0)
plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))

#Blending
# Alpha = 0.6
# Beta = 0.3

blended = cv2.addWeighted(src1=img, alpha=0.6, src2=img2, beta=0.3, gamma=0)
plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))
```

Smoothing or blurring

```python
img = cv2.imread('data/input/img/lenna.png')
#Gaussian blur dengan 7x7 kernel
blurred =cv2.GaussianBlur(img, (7,7), 0)
plt.imshow(cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB))
```

Image Thresholding

```python
img = cv2.imread('data/input/img/lenna.png')

#konversi gambar ke grayscale
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#binary tresholding
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(cv2.cvtColor(thresh1,cv2.COLOR_BGR2RGB))
```

```python
ret, thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(cv2.cvtColor(thresh2,cv2.COLOR_BGR2RGB))
```
